```python
# Opening JSON file
from data_ import MyDataset, coll
import json
import pandas as pd
from torch.utils.data import Dataset, DataLoader
f = open('train.json')  
# returns JSON object as 
# a dictionary
data1 = json.load(f)

kes=['nda-11', 'nda-16', 'nda-15', 'nda-10', 'nda-2', 'nda-1', 'nda-19', 'nda-12', 'nda-20', 'nda-3', 'nda-18', 'nda-7', 'nda-17', 'nda-8', 'nda-13', 'nda-5', 'nda-4']
rg = len(data1['documents'])
reference_list=[]
idx=0
for idx in range(rg):
  #Document level
  docid = data1['documents'][idx]['id']
  for ndas_key in kes:
    reference_list.append([docid, ndas_key,idx])
  idx=idx+1
ref_train = pd.DataFrame(reference_list ,columns = ['docid', 'nda_key', 'idx'])
```

```python
train_data=MyDataset(data1, ref_train, use_faiss = False)
train_loader=DataLoader(train_data,batch_size=20,collate_fn=coll, shuffle=False)
```

```python
k=1
for i, x in enumerate(train_loader):
  
  print(x['Label'].shape, x['Premise'].shape, x['Hypothesis'].shape)
  break
  k=k+1
  if k%100==0:
    print(k)
```
